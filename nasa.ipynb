```python
import csv
import numpy as np
import matplotlib.pyplot as plt


def read_csv_data(file_path, x_index=0, y_index=3):
    """Reads data from a CSV file and returns two lists: x (years) and y (averages)."""
    x, y = [], []
    
    # Read the CSV file
    with open(file_path, mode='r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            x.append(float(row[x_index]))
            y.append(float(row[y_index]))
    
    return x, y


def plot_data(x, y, model, prediction_range):
    """Plots the scatter plot of x and y data, and the polynomial regression line."""
    plt.scatter(x, y, label="CO2 Levels")
    plt.plot(prediction_range, model(prediction_range), color='red', label="Model Prediction")
    plt.xlabel("Year")
    plt.ylabel("CO2 Levels (ppm)")
    plt.title("CO2 Levels Over Time with Polynomial Regression")
    plt.legend()
    plt.show()

co2_years_copy = []
co2_average_copy = []

def main():
    # Read data from the CSV file
    file_path = 'data-co2.csv'
    years, average = read_csv_data(file_path)

    # Fit a polynomial regression model (degree 3)
    model = np.poly1d(np.polyfit(years, average, 3))

    # Generate a range of years for predictions
    prediction_range = np.linspace(min(years), 2050, 100)

    # Plot the data and the regression model
    plot_data(years, average, model, prediction_range)

    # Predict the CO2 level for 2050
    prediction_2050 = model(2050)
    print(f"Predicted CO2 level for 2050: {prediction_2050:.2f} ppm")

    for i in range(1950,2050):
        co2_years_copy.append(i)
        co2_average_copy.append(model(i))



if __name__ == "__main__":
    main()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt


def read_csv_data(file_path, x_index=0, y_index=2):
    """Reads data from a CSV file and returns two lists: x (years) and y (averages)."""
    x, y = [], []
    
    # Read the CSV file
    with open(file_path, mode='r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            x.append(float(row[x_index]))
            y.append(float(row[y_index]))
    
    return x, y


def plot_data(x, y, model, prediction_range):
    """Plots the scatter plot of x and y data, and the polynomial regression line."""
    plt.scatter(x, y, label="Temperature Anomalies")
    plt.plot(prediction_range, model(prediction_range), color='orange', label="Model Prediction")
    plt.xlabel("Year")
    plt.ylabel("Temperature Anomaly (C)")
    plt.title("Global Temperature Anomalies with Polynomial Regression")
    plt.legend()
    plt.show()

temperature_years_copy = []
temperature_average_copy = []

def main():
    # Read data from the CSV file
    file_path = 'temperature.csv'
    years, average = read_csv_data(file_path)

    # Fit a polynomial regression model (degree 3)
    model = np.poly1d(np.polyfit(years, average, 3))

    # Generate a range of years for predictions
    prediction_range = np.linspace(min(years), 2050, 100)

    # Plot the data and the regression model
    plot_data(years, average, model, prediction_range)

    # Predict the temperature anomaly for 2050
    prediction_2050 = model(2050)
    print(f"Predicted Temperature Anomaly for 2050: {prediction_2050:.2f} C")
    for i in range(1950,2050):
            temperature_years_copy.append(i)
            temperature_average_copy.append(model(i))


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Read the CSV file
file_path = 'newdata-sea-level.csv'  # Replace with the path to your file
sea_level_data = pd.read_csv(file_path)

# Step 2: Extract relevant columns: year and GMSL variation
years = sea_level_data['year+fraction_of_year'].values.reshape(-1, 1)
gmsl_variation = sea_level_data['GMSL_variation_'].values

# Step 3: Clean the data by removing NaN values
valid_data = ~np.isnan(gmsl_variation)
years_clean = years[valid_data]
gmsl_variation_clean = gmsl_variation[valid_data]

# Step 4: Fit a linear regression model
model = LinearRegression()
model.fit(years_clean, gmsl_variation_clean)

# Step 5: Predict sea level until 2050
future_years = np.arange(2024, 2051).reshape(-1, 1)
predicted_sea_levels = model.predict(future_years)

# Step 6: Plot the historical data and the predictions
plt.figure(figsize=(10, 6))
plt.scatter(years_clean, gmsl_variation_clean, label="Observed Sea Level", color='blue')
plt.plot(future_years, predicted_sea_levels, label="Predicted Sea Level (2024-2050)", color='red')
plt.xlabel('Year')
plt.ylabel('Sea Level Variation (mm)')
plt.title('Observed and Predicted Global Mean Sea Level Variation (GMSL)')
plt.legend()
plt.grid(True)
plt.show()
```
